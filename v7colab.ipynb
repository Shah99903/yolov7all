```python
import torch
import cv2
import numpy as np
import time
import glob
import os
from models.experimental import attempt_load
from utils.datasets import letterbox
from utils.general import non_max_suppression, scale_coords
from utils.torch_utils import select_device

# Path to the custom-trained weights
model_path = '/content/gdrive/MyDrive/v7_all/yolov7/runs/train/exp2/weights/best.pt'

# Path to the directory containing the test images
image_dir = '/content/gdrive/MyDrive/v7_all/yolov7/data/newyork-data-4/test/images'

# Path to save FPS and processing times
fps_output_dir = '/content/gdrive/MyDrive/v7_all/yolov7/runs/fps'
os.makedirs(fps_output_dir, exist_ok=True)

# Select Device
device = select_device('')  # Select GPU if available, otherwise CPU

# Load YOLOv7 Model
model = attempt_load(model_path, map_location=device)  # Load model
model.eval()  # Set model to evaluation mode

# Get list of image paths
image_paths = glob.glob(f"{image_dir}/*.jpg")

# Initialize variables to calculate average FPS and frame times
total_time = 0
num_images = 0

# Open a file to save FPS and processing times
fps_file_path = os.path.join(fps_output_dir, 'fps_and_times.txt')
with open(fps_file_path, 'w') as file:
    file.write('Image Path, Inference Time (s)\n')

    # Process each image
    for image_path in image_paths:
        # Prepare the Sample Image
        img = cv2.imread(image_path)

        if img is None:
            print(f"Failed to load image at {image_path}")
            continue

        img = letterbox(img, new_shape=(640, 640))[0]
        img = img[:, :, ::-1].transpose(2, 0, 1)  # Convert to RGB and format
        img = np.ascontiguousarray(img)
        img = torch.from_numpy(img).to(device).float()
        img /= 255.0  # Normalize to [0, 1]
        img = img.unsqueeze(0)  # Add batch dimension

        # Measure Inference Time
        start_time = time.time()
        with torch.no_grad():
            pred = model(img)[0]
            pred = non_max_suppression(pred, 0.25, 0.45, classes=None, agnostic=False)
        end_time = time.time()

        # Calculate time taken for inference
        inference_time = end_time - start_time
        total_time += inference_time
        num_images += 1

        # Write the image path and inference time to the file
        file.write(f"{image_path}, {inference_time:.3f}\n")

        print(f"Inference time for {image_path}: {inference_time:.3f} seconds")

# Calculate Average FPS
if num_images > 0:
    average_inference_time = total_time / num_images
    fps = 1 / average_inference_time
    print(f"Processed {num_images} images")
    print(f"Average Inference Time: {average_inference_time:.3f} seconds")
    print(f"Average FPS: {fps:.2f}")

    # Save the average FPS to the file
    with open(fps_file_path, 'a') as file:
        file.write(f"Average FPS: {fps:.2f}\n")
else:
    print("No images were processed.")
```
